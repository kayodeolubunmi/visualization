Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT *
FROM master"""
```

```python
df = pd.read_sql(QUERY, conn)
```

```python
plt.figure(figsize=(10, 6))
sns.distplot(df['ConvertedComp'], bins=30, kde=False)  
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'], vert=False)
plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

plt.scatter(df['Age'], df['WorkWeekHrs'])
plt.title('Scatterplot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

plt.figure(figsize=(10, 6))
bubble = sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(20,2000), alpha=0.5, color='blue', data=df)
plt.title('Bubble Plot of WorkWeekHrs, CodeRevHrs and Age')
plt.xlabel('Work Week Hours')
plt.ylabel('CodeRevHrs')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

age_group = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
```

```python
fig, ax = plt.subplots()
age_group.set_index('Age')[['WorkWeekHrs', 'CodeRevHrs']].plot(kind='bar', stacked=True, ax=ax)
ax.set_title('Stacked Bar Chart ofWorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
ax.set_xlabel('Age')
ax.set_ylabel('Values')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
age_group_new = df[(df['Age'] >= 45) & (df['Age'] <= 60)]
```

```python
median_ConvertedComp= age_group_new .groupby('Age')['ConvertedComp'].median()
```

```python
fig, ax = plt.subplots()
median_ConvertedComp.plot(kind='line', marker='o', ax=ax)
ax.set_title('Median ConvertedComp for Age Group 45 to 60')
ax.set_xlabel('Age')
ax.set_ylabel('Median ConvertedComp')
plt.show()
```
